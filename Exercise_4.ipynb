```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and select the "R" and "J" columns
df = pd.read_excel('exact.xlsx')
df = df.loc[:, ["R", "J"]]

# Create a list of 1000 x-values from 1 to 1000
x_values = list(range(1, 1001))

plt.plot(x_values, df["R"], color="orange", label="R")
plt.plot(x_values, df["J"], color="green", label="J")

plt.title("Raw data")
plt.legend()
plt.show()
```

## Denoise Data
The Savitzky-Golay filter is a type of digital filter used for smoothing data in the time domain, as well as for differentiating and integrating signals. It is particularly useful for removing noise from data, especially when the noise is high frequency and the signal has a smooth underlying trend.

The filter works by fitting a polynomial function to a sliding window of adjacent data points, and then using this polynomial to estimate the value of the central data point. The degree of the polynomial and the width of the sliding window are parameters that can be adjusted to control the level of smoothing and the degree of noise reduction.

One of the advantages of the Savitzky-Golay filter over other smoothing techniques, such as moving averages or median filters, is that it preserves the shape of the original signal to a greater extent, while still effectively removing noise. This is because the filter uses a polynomial function to model the trend of the signal, rather than simply averaging or interpolating neighboring data points.

Overall, the Savitzky-Golay filter is a powerful tool for denoising and smoothing data, and has many applications in fields such as signal processing, image processing, and data analysis.

```python
from scipy.signal import savgol_filter

denoised_R = savgol_filter(df["R"], window_length=200, polyorder=3)
plt.plot(x_values, denoised_R, color="orange", lw=3, label="R denoised")

denoised_J = savgol_filter(df["J"], window_length=200, polyorder=3)
plt.plot(x_values, denoised_J, color="green", lw=3, label="J denoised")

plt.legend() 
plt.title("Savitzky-Golay Filter")
plt.show()
```
